```python
prices = yf.download("AAPL", start="2024-01-01", end="2024-04-01")
market_days = prices.index.normalize().strftime("%Y-%m-%d").tolist()
prices = prices["Close"]
prices.index = prices.index.strftime("%Y-%m-%d")  # make it string for safe matching

print(prices)
capital = 10000
position = 0
shares_per_trade = 50
portfolio = []
previous_day = market_days[0]

for decision in decisions:
    date = decision["date"]
    execution_price = prices.loc[previous_day, "AAPL"]
    close_price = prices.loc[date, "AAPL"]

    if decision["decision"].lower().startswith("buy"):
        if position <0:
            capital -= abs(position) * execution_price
            position=0
        if position == 0:    
            position = shares_per_trade
            capital -= position * execution_price
    elif decision["decision"].lower().startswith("sell"):
        if position >0 :
            capital += position * execution_price
            position=0
        if position == 0:
            position = -shares_per_trade
            capital += abs(position) * execution_price

    equity = capital + position * close_price
    portfolio.append({"date": date, "equity": equity})
    previous_day = date

df = pd.DataFrame(portfolio)
print(df)
df["date"] = pd.to_datetime(df["date"])
```

```python
buy_hold_portfolio = []
initial_price = prices.loc[market_days[0], "AAPL"]  # Buy at first decision date's close
shares_held = 10000 // initial_price
remaining_cash = 10000 - shares_held * initial_price
for date in market_days[1:]:
    close_price = prices.loc[date, "AAPL"]
    buy_hold_portfolio.append({"date": date, "equity": remaining_cash+shares_held*close_price})
df2 = pd.DataFrame(buy_hold_portfolio)
df2["date"] = pd.to_datetime(df2["date"])
```

```python
plt.plot(df["date"], df["equity"], label = 'LLM Strategy')
plt.plot(df2["date"], df2["equity"], label = 'Buy and Hold')
plt.title("Equity Curve")
plt.xlabel("Date")
plt.ylabel("Portfolio Value")


plt.legend()
plt.grid()
plt.show()
```
